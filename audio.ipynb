```python
import pyttsx3
import PyPDF2
book = open("oop.pdf", "rb") #the rb is for reading as a binary book

pdfreader = PyPDF2.PdfFileReader(book)
pages = pdfreader.numPages
print(pages) # to check how many pages there are in the pdf file

speaker = pyttsx3.init()
voices = engine.getProperty('voices')
#rate = engine.getProperty('rate')
for voice in voices:
    #print(voice.id)
    if voice.languages[0] == b'\x02en-gb':
        engine.setProperty('voice', voice.id)
        break

speaker.setProperty('rate', 160)
page = pdfreader.getPage(7) #to ask it to read from a certain page, for example here, from page 8 (started at 0)
#extract the text from the above:
text = page.extractText()
speaker.say(text)
speaker.runAndWait()
speaker.close()
```

```python
#there doesn't seem to be any female voice on linux... might work on windows, via command prompt.
for voice in voices:
  print("ID: %s" % voice.id)
  print("Name: %s" % voice.name)
  print("Age: %s" % voice.age)
  print("Gender: %s" % voice.gender)
  print("Languages Known: %s" % voice.languages)
```
